Data Analysis:

Firstly, we download our dataset from website, store it to our data folder and read it by using read_excel function. Since our dataset is already tidy enough, the only thing we did is to sort our UNS variables by four different levels.This give us a clean and tidy dataset.

```r
library(readxl)
library(repr)
library(dplyr)
library(tidyverse)

url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/00257/Data_User_Modeling_Dataset_Hamdi%20Tolga%20KAHRAMAN.xls"
library(tidyverse)

# your code here
download.file(url <- "https://archive.ics.uci.edu/ml/machine-learning-databases/00257/Data_User_Modeling_Dataset_Hamdi%20Tolga%20KAHRAMAN.xls", destfile = "data/Data_User_Modeling_Dataset_Hamdi%20Tolga%20KAHRAMAN.xls")
user_knowledge <- read_excel(path = "data/Data_User_Modeling_Dataset_Hamdi%20Tolga%20KAHRAMAN.xls",sheet ="Training_Data",range = cell_cols("A:F"))

user_knowledge <- user_knowledge %>%
    mutate(UNS = factor(UNS, levels = c('very_low','Low','Middle','High'))) %>%
    arrange(UNS)
user_knowledge
```

As we mentioned above, our target value is UNS. So we want to get more information about this variables.

```r
count_uns <- user_knowledge %>%
    group_by(UNS) %>%
    count(UNS)

count_bar_chart <- ggplot(count_uns, aes(x = UNS, y = n)) + 
    geom_bar(stat = "identity") +
    ggtitle("UNS distribution") +
    xlab("UNS level") +
    ylab("total number of user in different UNS level")
count_bar_chart
```

We create a bar graph to present the distribution of UNS. From the above graph, we can see that the level of UNS are quite normal distributed.This means that the middle of UNS has highest distribution and a relatively low distribution from side bar.This information might help us for futher analysis.

Then, we want to see if there is a linear relationship between each of the variable with our target value(UNS).

```r
STG <- ggplot(user_knowledge, aes(x = STG, y = UNS)) +
    geom_point()  +
    theme(text = element_text(size = 20))
STG

SCG <- ggplot(user_knowledge, aes(x = SCG, y = UNS)) +
    geom_point() +
    theme(text = element_text(size = 20))
SCG

STR <- ggplot(user_knowledge, aes(x = STR, y = UNS)) +
    geom_point() +
    theme(text = element_text(size = 20))
STR

LPR <- ggplot(user_knowledge, aes(x = LPR, y = UNS)) +
    geom_point()  +
    theme(text = element_text(size = 20))
LPR

PEG <- ggplot(user_knowledge, aes(x = PEG, y = UNS)) +
    geom_point()  +
    theme(text = element_text(size = 20))
PEG
```

From the above plot, we can see that only PEG has a strong linear relationship with UNS. Also to be noticed that in very_low level of UNS, most of PEG are around 0 to 0.125. In Low level of UNS, most of PEG are around 0.125 to 0.375.In Middle level of UNS, most of PEG are around 0.5 to 0.65.In High level of UNS, most of PEG are around 0.6 to 0.9. This means that this four part of UNS are subgroup and they are quite indenpendent. This means that PEG could be a great predictor variable for our futher classification model.
